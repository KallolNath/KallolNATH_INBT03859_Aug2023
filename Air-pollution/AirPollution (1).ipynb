# **Introduction**

The Beijing PM2.5 Data Data Set is a comprehensive collection of air pollution data specifically focused on PM2.5 (particulate matter with a diameter of 2.5 micrometers or smaller) levels in Beijing, China. PM2.5 is considered a major air pollutant and has significant impacts on human health and the environment.

This dataset provides information on daily PM2.5 concentrations recorded at various monitoring stations across Beijing. The data set covers a considerable time period and includes measurements taken from multiple stations within the city. Each data point typically includes the date, time, and PM2.5 concentration level, along with other relevant meteorological and environmental factors.

The purpose of this data set is to monitor and analyze the air quality in Beijing over time, identify trends, and assess the effectiveness of pollution control measures. It serves as a valuable resource for researchers, scientists, policymakers, and environmentalists interested in studying air pollution, its sources, and its impacts on public health.

Researchers can utilize this dataset to conduct various analyses, such as investigating seasonal variations in PM2.5 levels, identifying pollution hotspots within the city, assessing the correlation between PM2.5 and weather conditions, and evaluating the effectiveness of pollution reduction strategies implemented by authorities.

By examining this data set, one can gain insights into the patterns and trends of PM2.5 pollution in Beijing, helping to inform policy decisions aimed at improving air quality and protecting the health and well-being of the city's residents.

## **Model Building**

```python
## Putting the target value in the end

ap = ap.iloc[ : , [0,1,2,3,4,6,7,8,9,10,11,12,5]]
ap.info()
```

```python
## splitting the data into training and testing set

from sklearn.model_selection import train_test_split

X = ap.drop(columns = ["pm2.5"], axis=1)
Y = ap["pm2.5"]

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## **Ridge Regression**

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV # to find the best parameters
from sklearn import metrics

ridge_regressor = Ridge()
```

```python
## Finding the best value of alpha with the help of gridsearchcv

parameters = {'alpha':[1,2,3,4,5,6,7,8,9]}
ridgecv=GridSearchCV(ridge_regressor,parameters,scoring='neg_mean_squared_error',cv=5)
ridgecv.fit(x_train,y_train)
```

```python
## Printing the best value of alpha
print(ridgecv.best_params_)
```

```python
## Building the model

ridge = Ridge(alpha=1)
ridge.fit(x_train,y_train)
ridge_y_train_predict = ridge.predict(x_train)
ridge_y_test_predict = ridge.predict(x_test)
```

## **Random Forest Regression**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
## Finding the best parameter using randomsearchcv

from sklearn.model_selection import RandomizedSearchCV
param_grid = { "n_estimators" : [50,100,150,200,250],
              "max_features" : [0.6,1.0,"sqrt"],
               "max_depth" : [2,3,5,7],
               "max_samples" : [0.5,0.6,0.75,1.0],
               "min_samples_split" : [2,3,4,5,6,7,8,9],
               "min_samples_leaf" : [2,3,4,5,6,7,8,9],
               "bootstrap" : [True,False]
               }

rf_grid = RandomizedSearchCV(estimator = rf,
                       param_distributions = param_grid,
                       cv = 5,
                       verbose=2,
                       n_jobs = -1)
```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf = RandomForestRegressor(n_estimators = 250,
                           max_depth = 7,
                           min_samples_split = 2,
                           min_samples_leaf =7,
                           max_samples=0.75,
                           bootstrap=True)

rf.fit(x_train,y_train)
rf_predict_y_train = rf.predict(x_train)
rf_predict_y_test = rf.predict(x_test)
```

## **Model Evaluation**

### **For Ridge Regression**

```python
# Performance evaluation of training data for Ridge Regression

print("Training Model Performance of Ridge Regression \n")

print("Mean absolute error      :",metrics.mean_absolute_error(y_train,ridge_y_train_predict))
print("Mean squared error       :",metrics.mean_squared_error(y_train,ridge_y_train_predict))
print("Root mean squared error  :",np.sqrt(metrics.mean_squared_error(y_train,ridge_y_train_predict)))
print("R2 Score                 :",metrics.r2_score(y_train,ridge_y_train_predict))
```

```python
# Performance evaluation of testing data for Ridge Regression

print("Testing Model Performance of Ridge Regression \n")

print("Mean absolute error      :",metrics.mean_absolute_error(y_test,ridge_y_test_predict))
print("Mean squared error       :",metrics.mean_squared_error(y_test,ridge_y_test_predict))
print("Root mean squared error  :",np.sqrt(metrics.mean_squared_error(y_test,ridge_y_test_predict)))
print("R2 Score                 :",metrics.r2_score(y_test,ridge_y_test_predict))
```

### **For Random Forest**

```python
# Performance evaluation of training data for Random Forest Regression

print("Training Model Performance of Random Forest Regression \n")

print("Mean absolute error      :",metrics.mean_absolute_error(y_train,rf_predict_y_train ))
print("Mean squared error       :",metrics.mean_squared_error(y_train,rf_predict_y_train ))
print("Root mean squared error  :",np.sqrt(metrics.mean_squared_error(y_train,rf_predict_y_train )))
print("R2 Score                 :",metrics.r2_score(y_train,rf_predict_y_train ))
```

```python
# Performance evaluation of testing data for Random Forest Regression

print("Testing Model Performance of Random Forest Regression \n")

print("Mean absolute error      :",metrics.mean_absolute_error(y_test,rf_predict_y_test))
print("Mean squared error       :",metrics.mean_squared_error(y_test,rf_predict_y_test))
print("Root mean squared error  :",np.sqrt(metrics.mean_squared_error(y_test,rf_predict_y_test)))
print("R2 Score                 :",metrics.r2_score(y_test,rf_predict_y_test))
```

## **Cross Validation and Overfitting Check**

```python
# Importing Cross_val_score and K-Fold from sklearn

from sklearn.model_selection import cross_val_score, KFold
cvs = cross_val_score
k_fold = KFold(n_splits = 10, shuffle = True, random_state = 42)
```

```python
# Checking if Ridge Model is overfitted or not

train_score = ridge.score(x_train,y_train)
test_score = ridge.score(x_test,y_test)

if train_score > test_score:
  print("Ridge Regression Model is Overfitted")
  print("Train Score: ",train_score)
  print("Test Score: ",test_score)

else:
  print("Ridge Regression is not Overfitted")
  print("Train Score: ",train_score)
  print("Test Score: ",test_score)
```

```python
# Since the Ridge Regression Model is overfitted, we will do k-fold cross validation test.

ridge_score = cvs(ridge,x_test,y_test,cv=k_fold)
ridge_score
```

```python

kf = KFold(n_splits = 10, shuffle = True, random_state = 42)

test_score = []
train_score = []

X = pd.DataFrame(X)
Y = pd.DataFrame(Y)
# Performing Cross Validation
for train_index, test_index in kf.split(X):
  x_train, x_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]


  ridge.fit(x_train,y_train)

  y_pred_test = ridge.predict(x_test)
  y_pred_train = ridge.predict(x_train)

  mae_test = metrics.mean_absolute_error(y_pred_test,y_test)
  mae_train = metrics.mean_absolute_error(y_pred_train,y_train)

  test_score.append(mae_test)
  train_score.append(mae_train)

# Calculating and printing MAE score of each training fold
print("MAE Score for each Fold in Training \n")
for fold, score in enumerate(train_score):
  print(f"Fold {fold+1} : {score}")

# Calculating and printing MAE score of each testing fold
print("\n MAE Score for each Fold in Testing \n")
for fold, score in enumerate(test_score):
  print(f"Fold {fold+1} : {score}")


# Average MAE Score of Training Set
avg_MAE_train = sum(train_score) / len(train_score)
print("\n Average MAE Train Score: ", avg_MAE_train)

# Average MAE Score of Testing Set
avg_MAE_test = sum(test_score) / len(test_score)
print("\n Average MAE Test Score: ", avg_MAE_test)

# Minimum MAE Score of Training Set
min_MAE_train = min(train_score)
print("\n Minimum MAE Train Score: ", min_MAE_train)

# Minimum MAE Score of Testing Set
min_MAE_test = min(test_score)
print("\n Minimum MAE Test Score: ", min_MAE_test)

# Maximum MAE Score of Training Set
max_MAE_train = max(train_score)
print("\n Maximum MAE Train Score: ", max_MAE_train)

# Maximum MAE Score of Testing Set
max_MAE_test = max(test_score)
print("\n Maximum MAE Test Score: ", max_MAE_test)
```
