# Introduction

The Heart-Disease-Dataset is a comprehensive repository containing a total of 76 attributes, with specific focus placed on a subset of 14 attributes consistently utilized in a range of published experiments. This dataset has catched significant attention within the machine learning research community, particularly due to the importance of the Cleveland database, which is featured in various studies. The pivotal attribute referred to as "Target" plays a central role, serving as an indicator of the presence or absence of heart disease in individual patients. This attribute has integer value 0 and 1, wherein a value of 0 signifies the absence of heart disease, while values 1 denote presence of heart desiease.

## **Train-Test Split**

```python
# Importing train-test split from sklearn
from sklearn.model_selection import train_test_split

X = Heart_disease.drop(columns = ["target"], axis =1)
Y = Heart_disease["target"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## **Building Algorithm**

### **KNN Classifier**

```python
# Importing KNN Classifier from sklearn
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors= 5)
knn.fit(x_train, y_train)
knn_pred_y_train = knn.predict(x_train)
knn_pred_y_test = knn.predict(x_test)
```

### **Naive Baye's**

```python
# importing GaussianNB from sklearn for Naive Bayes classification
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train, y_train)
nb_pred_y_train = nb.predict(x_train)
nb_pred_y_test = nb.predict(x_test)
```

### **Decision Tree Classifier**

```python
# Importing Decision Tree Classifier from sklearn
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt_pred_y_train = dt.predict(x_train)
dt_pred_y_test = dt.predict(x_test)
```

## **Model Evaluation**

### **For KNN**

```python
# Importing different matrics from sklearn to evaluate the model

from sklearn.metrics import recall_score,precision_score,accuracy_score,f1_score,roc_auc_score,classification_report,confusion_matrix


print("Model Performance of KNN Classifier \n")

# Performance evaluation of training data
print("----Training Performance---- \n")
print("Accuracy        :",accuracy_score(y_train,knn_pred_y_train))
print("Precision       :",precision_score(y_train,knn_pred_y_train))
print("Recall          :",recall_score(y_train,knn_pred_y_train))
print("f1 Score        :",f1_score(y_train,knn_pred_y_train))
print("ROC AUC Score   :",roc_auc_score(y_train,knn_pred_y_train))
print("Confusion Matrix:\n",confusion_matrix(y_train,knn_pred_y_train))
print("\nClassification Matrix :\n",classification_report(y_train,knn_pred_y_train))
```

```python
# Performance Evaluation of Testing Data

print("----Testing Performance---- \n")
print("Accuracy        :",accuracy_score(y_test,knn_pred_y_test))
print("Precision       :",precision_score(y_test,knn_pred_y_test))
print("Recall          :",recall_score(y_test,knn_pred_y_test))
print("f1 Score        :",f1_score(y_test,knn_pred_y_test))
print("ROC AUC Score   :",roc_auc_score(y_test,knn_pred_y_test))
print("Confusion Matrix:\n",confusion_matrix(y_test,knn_pred_y_test))
print("\nClassification Matrix :\n",classification_report(y_test,knn_pred_y_test))
```

```python
# Test accuracy of 81 is pretty good but lets find the best value of k using cross validation

from sklearn.model_selection import cross_val_score

k_values = [i for i in range (1,31)]
scores = []

scaler = StandardScaler()
X = scaler.fit_transform(X)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, X, Y, cv=5)
    scores.append(np.mean(score))

# We can plot the result in a Line Plot

sns.lineplot(x = k_values, y = scores, marker = 'o')
plt.xlabel("K Values")
plt.ylabel("Accuracy Score")
```

Since 1, 2 and 3 gives the best accuracy score. So we can go with 3 as having a 1 and 2 as value of K doesn't really make sense.

```python
from sklearn import metrics

kf = KFold(n_splits = 10, shuffle = True, random_state = 42)

test_score = []
train_score = []

X = pd.DataFrame(X)
Y = pd.DataFrame(Y)
# Performing Cross Validation
for train_index, test_index in kf.split(X):
  x_train, x_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]


  knn.fit(x_train,y_train)

  y_pred_test = knn.predict(x_test)
  y_pred_train = knn.predict(x_train)

  accuracy_test = metrics.accuracy_score(y_pred_test,y_test)
  accuracy_train = metrics.accuracy_score(y_pred_train,y_train)

  test_score.append(accuracy_test)
  train_score.append(accuracy_train)

# Calculating and printing accuracy score of each training fold
print("Accuracy Score for each Fold in Training \n")
for fold, score in enumerate(train_score):
  print(f"Fold {fold+1} : {score}")

# Calculating and printing accuracy score of each testing fold
print("\nAccuracy Score for each Fold in Testing \n")
for fold, score in enumerate(test_score):
  print(f"Fold {fold+1} : {score}")


# Average Accuracy Score of Training Set
avg_accuracy_train = sum(train_score) / len(train_score)
print("\n Average Accuracy Train Score: ", avg_accuracy_train)

# Average Accuracy Score of Testing Set
avg_accuracy_test = sum(test_score) / len(test_score)
print("\n Average Accuracy Test Score: ", avg_accuracy_test)

# Minimum Accuracy Score of Training Set
min_accuracy_train = min(train_score)
print("\n Minimum Accuracy Train Score: ", min_accuracy_train)

# Minimum Accuracy Score of Testing Set
min_accuracy_test = min(test_score)
print("\n Minimum Accuracy Test Score: ", min_accuracy_test)

# Maximum Accuracy Score of Training Set
max_accuracy_train = max(train_score)
print("\n Maximum Accuracy Train Score: ", max_accuracy_train)

# Maximum Accuracy Score of Testing Set
max_accuracy_test = max(test_score)
print("\n Maximum Accuracy Test Score: ", max_accuracy_test)

```
